```python
import matplotlib.pyplot as plt

def make_plot(scores_baseline, scores_baseline_z, scores_ct, scores_ct_z, scores_mri, scores_mri_z):
# Prepare the order of models from liver_scores_ct dictionary
    ordered_models = list(scores_ct.keys())

    # Sum the values for each dictionary
    def sum_scores(scores_dict):
        return {model: sum(values) / 2 for model, values in scores_dict.items()}

    # Compute summed scores for each dictionary
    sum_scores_baseline = sum_scores(scores_baseline)
    sum_scores_baseline_z = sum_scores(scores_baseline_z)
    sum_scores_ct = sum_scores(scores_ct)
    sum_scores_ct_z = sum_scores(scores_ct_z)
    sum_scores_mri = sum_scores(scores_mri)
    sum_scores_mri_z = sum_scores(scores_mri_z)

    # List of dictionaries and corresponding labels
    scores_dicts = [sum_scores_baseline, sum_scores_baseline_z, sum_scores_ct, sum_scores_ct_z, sum_scores_mri, sum_scores_mri_z]
    conditions = ['Baseline (1)', 'Baseline (2)', 'CT \u2192 MRI (1) ', 'CT \u2192 MRI (2)',  'MRI \u2192 CT (1)', 'MRI \u2192 CT (2)']
    #markers = ['o', 's', 'v', '^', 'P', 'X']
    markers = ['s', 's', '^', '^', 'o', 'o']

    # Plotting
    plt.figure(figsize=(8, 5))

    for i, scores in enumerate(scores_dicts):
        scores_values = [scores.get(model, None) for model in ordered_models]  # Use None for missing models
        plt.plot(ordered_models, scores_values, label=conditions[i], linestyle='', marker=markers[i], markersize=8)

    # Customize the plot
    plt.xlabel('Models')
    plt.ylabel('Mean DSC Scores')
    plt.title('')
    plt.xticks(rotation=45)
    #plt.legend(loc='lower right', ncol=2) #for liver
    plt.legend()
    plt.grid(axis='y')
    plt.tight_layout()
    plt.savefig("model_scores_plot.png", format="png", dpi=300)
    plt.show()
```

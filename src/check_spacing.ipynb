```python
import os
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt

# Specify the directory containing the images
image_dir = '/lsdf/data/medical/large_scale/AMOS/AMOS/amos22/imagesTr/'

# Lists to hold the zooms for each axis
zooms_x = []
zooms_y = []
zooms_z = []

# Iterate over all files in the directory
for filename in os.listdir(image_dir):
    # Check if the file is a NIfTI file
    if filename.endswith('.nii') or filename.endswith('.nii.gz'):
        # Construct the full path to the image file
        image_path = os.path.join(image_dir, filename)
        
        # Load the image using nibabel
        img = nib.load(image_path)
        
        # Get the zooms (voxel spacings) for the image
        zooms = img.header.get_zooms()
        
        # Append the zooms to the respective axis lists
        if len(zooms) >= 3:
            zooms_x.append(zooms[0])  # First axis (X)
            zooms_y.append(zooms[1])  # Second axis (Y)
            zooms_z.append(zooms[2])  # Third axis (Z)
        elif len(zooms) == 2:
            zooms_x.append(zooms[0])  # X
            zooms_y.append(zooms[1])  # Y
            zooms_z.append(zooms[1])  # Z (assume equal spacing for 2D)
        elif len(zooms) == 1:
            zooms_x.append(zooms[0])  # X
            zooms_y.append(zooms[0])  # Y
            zooms_z.append(zooms[0])  # Z (assume equal spacing)

# Convert lists of zooms to NumPy arrays for easier handling
zooms_x_array = np.array(zooms_x)
zooms_y_array = np.array(zooms_y)
zooms_z_array = np.array(zooms_z)

# Define font sizes
title_fontsize = 16
label_fontsize = 14
tick_fontsize = 12

# Create histograms for each axis
plt.figure(figsize=(11, 3))

# Histogram for X axis
plt.subplot(1, 3, 1)
plt.hist(zooms_x_array, bins=30, alpha=0.75, color='blue', edgecolor='black')
plt.title('Histogram of Voxel Spacing (X axis)')
plt.xlabel('Voxel Spacing (mm)')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)

# Histogram for Y axis
plt.subplot(1, 3, 2)
plt.hist(zooms_y_array, bins=30, alpha=0.75, color='green', edgecolor='black')
plt.title('Histogram of Voxel Spacing (Y axis)')
plt.xlabel('Voxel Spacing (mm)')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)

# Histogram for Z axis
plt.subplot(1, 3, 3)
plt.hist(zooms_z_array, bins=30, alpha=0.75, color='red', edgecolor='black')
plt.title('Histogram of Voxel Spacing (Z axis)')
plt.xlabel('Voxel Spacing (mm)')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
